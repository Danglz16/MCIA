```python
import numpy as np
import pandas as pd
import sympy as sp
import matplotlib.pyplot as plt
```

Partiendo de la ecuacion:  
z = $x^3+x^2+y^3+y^2$  
z = $\theta_0^3+\theta_0^2+\theta_1^3+\theta_1^2$



Obtenemos la derivada parcial de X y de Y con lo cual obtenemos:  

$3x^2 + 2x$  
$3y^2 + 2y$


```python
def func(x, y):
    return x**3 + x**2 + y**3 + y**2

def dparc_x (x, y):
    return 3*x**2 + 2*x
    
def dparc_y(x, y):
    return 3*y**2 + 2*y
```

```python
#def graph_3d(xs, ys, zs, hor, vert):
def graph_3d(xs, ys, zs):
    from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    X = np.linspace(-1, 1, 100)
    Y = np.linspace(-1, 1, 100)
    X,Y = np.meshgrid(X, Y)
    Z = X**3 + X**2 + Y**3 + Y**2
    ax.plot_surface(X, Y, Z, cmap='plasma', linewidth=0, antialiased=True, alpha=0.5)
    ax.scatter(xs, ys, zs, color='r', s=5)  
    #ax.view_init(elev=vert, azim=hor, roll=0)          # View angle
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z = x^3 + x^2 + y^3 + y^2')
    ax.set_title('Superficie 3D')
    plt.show()
```

```python
#def grad (theta1, theta2, alfa, epochs, hor, vert):
def grad (theta1, theta2, alfa, epochs):
    xs = []
    ys = []
    zs = []
    for i in range(epochs):
        theta1 = theta1-alfa*dparc_x(theta1,theta2)
        theta2 = theta2-alfa*dparc_y(theta1,theta2)

        xs.append(theta1)
        ys.append(theta2)
        zs.append(func(theta1, theta2))

    #graph_3d(xs, ys, zs, hor, vert)
    graph_3d(xs, ys, zs)
    return theta1, theta2
```

# Gradiente Automatica

```python
def grad_linreg_stable(X, y, alpha=0.05, epochs=2000, theta0=0.0, theta1=0.0, return_history=False):
    X = np.asarray(X, dtype=float).ravel()
    y = np.asarray(y, dtype=float).ravel()
    m = X.size

    # 1) Escalado (estandarización)
    mux, sx = X.mean(), X.std(ddof=0)
    muy, sy = y.mean(), y.std(ddof=0)

    Xs = (X - mux) / sx
    ys = (y - muy) / sy

    t0, t1 = 0.0, 0.0  # Espacio escalado (ignora args iniciales)
    hist = {"t0":[], "t1":[], "J":[]}

    for _ in range(epochs):
        yhat_s = t1*Xs + t0
        res = yhat_s - ys

        # 2) Gradientes en espacio escalado
        g0 = np.mean(res)
        g1 = np.mean(res * Xs)

        # 3) Gradient clipping para estabilidad
        g0 = np.clip(g0, -1e3, 1e3)
        g1 = np.clip(g1, -1e3, 1e3)

        # 4) Actualización
        t0 -= alpha * g0
        t1 -= alpha * g1

        if return_history:
            J = 0.5*np.mean(res**2)
            hist["t0"].append(t0); hist["t1"].append(t1); hist["J"].append(J)

    # 5) Volver a la escala original
    # y_s = t1*x_s + t0  -> y = sy*y_s + muy; x_s=(x-mux)/sx
    theta1_orig = (sy/sx) * t1
    theta0_orig = muy + sy*t0 - theta1_orig * mux

    if return_history:
        return theta0_orig, theta1_orig, hist
    return theta0_orig, theta1_orig
```

```python
df = pd.read_csv('DB/Advertising.csv')
df.info()
```

```python
theta0, theta1, hist = grad_linreg_stable(df["TV"].values, df["Sales"].values,
                                          alpha=0.05, epochs=5000, return_history=True)
print("θ0 =", theta0, "  θ1 =", theta1)
```

```python
plt.scatter(df["TV"], df["Sales"], color="blue", label="Datos reales")


x_line = np.linspace(df["TV"].min(), df["TV"].max(), 100)
y_line = theta1 * x_line + theta0
plt.plot(x_line, y_line, color="red", linewidth=2, label="Recta ajustada")

plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Regresión Lineal con Gradiente Descendente")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_pred = theta1 * df["TV"].values + theta0

# Cálculo del Error Medio Cuadrático
EMC = np.mean((df["Sales"].values - y_pred)**2)
print("Error Medio Cuadrático (EMC):", EMC)
```

```python
plt.scatter(df["TV"], df["Sales"], color="blue", label="Datos reales")

plt.plot(x_line, y_line, color="red", linewidth=2,
         label=f"Recta ajustada (EMC={EMC:.2f})")

plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Regresión Lineal con Gradiente Descendente")
plt.legend()
plt.grid(True)
plt.show()
```

# N-Dimensiones

```python
def grad_multivariada(X, y, alpha=0.01, epochs=2000, return_history=False):
    """
    X: matriz de características (m muestras x n variables)
    y: vector de salida (m)
    """
    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float).reshape(-1, 1)
    m, n = X.shape

    # Inicializar theta (n x 1)
    theta = np.zeros((n, 1))
    hist = {"theta": [], "J": []}

    for _ in range(epochs):
        y_pred = X @ theta
        error = y_pred - y
        grad = (1/m) * (X.T @ error)
        theta -= alpha * grad

        if return_history:
            J = (1/(2*m)) * np.sum(error**2)
            hist["theta"].append(theta.copy())
            hist["J"].append(J)

    if return_history:
        return theta, hist
    return theta
```

```python
# Supón que quieres usar TV, Radio y Newspaper como variables independientes:

X_multi = df[["TV", "Radio", "Newspaper"]].values # Características
X_multi = np.hstack([np.ones((X_multi.shape[0], 1)), X_multi])  # Agregar Columna de 1 para sesgo
y = df["Sales"].values # Salida

# Aplicar Gradiente Descendente
theta_multi, hist = grad_multivariada(X_multi, y, alpha=0.00001, epochs=500, return_history=True)
print("Theta multivariada:", theta_multi.ravel())

# Predicción
y_pred_multi = X_multi @ theta_multi

# Error Medio Cuadrático
EMC_multi = np.mean((y - y_pred_multi.ravel())**2)
print("Error Medio Cuadrático (EMC):", EMC_multi)
```

```python
plt.figure(figsize=(8,4))
plt.plot(hist["J"])
plt.xlabel("Época")
plt.ylabel("Función de costo J")
plt.title("Evolución del error durante el entrenamiento")
plt.grid(True)
plt.show()
```

### Espacio 3D con plano de descenso de gradiente

```python
from mpl_toolkits.mplot3d import Axes3D

# Selecciona solo dos variables
X_2 = df[["TV", "Radio"]].values
X_2 = np.hstack([np.ones((X_2.shape[0], 1)), X_2])
y = df["Sales"].values

# Entrena el modelo
theta_2, hist_2 = grad_multivariada(X_2, y, alpha=0.00001, epochs=1000, return_history=True)

# Malla para la superficie
tv_range = np.linspace(df["TV"].min(), df["TV"].max(), 30)
radio_range = np.linspace(df["Radio"].min(), df["Radio"].max(), 30)
TV_grid, Radio_grid = np.meshgrid(tv_range, radio_range)
Z_grid = (theta_2[0] + theta_2[1]*TV_grid + theta_2[2]*Radio_grid)

# Gráfica 3D
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["TV"], df["Radio"], df["Sales"], color='red', label='Datos reales')
ax.plot_surface(TV_grid, Radio_grid, Z_grid, color='green', alpha=0.5, label='Superficie ajustada')
ax.set_xlabel("TV")
ax.set_ylabel("Radio")
ax.set_zlabel("Sales")
ax.set_title("Regresión Lineal Multivariada (2 variables)")
plt.show()
```
